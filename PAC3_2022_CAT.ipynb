###Exercici 2: Agrupament amb algorismes genètics [4 punts]

El primer que s'ha de fer per implementar un algorisme genètic és decidir cóm codificar una solució del problema. En un problema d'agrupament, codificarem una solució a partir dels centroides dels diferents clusters que volem trobar.
Específicament, una solució serà un vector filera de dimensions $1 \times d*k $ on $k$ és el nombre de clusters que volem identificar en les dades i $d$ és la dimensionalitat de les dades (nombre d'atributs). Com que en el nostre cas les dades que hem generat tenen d=2 i k=3, procedirem a fer servir aquests valors durant tot l'exercici i la codificació d'una solució $S$ vindrà donada pel vector de dimensions $1 \times 6$

$S = (cx_1, cy_1, cx_2, cy_2, cx_3, cy_3)$

on $(cx_i,cy_i)$ representen les coordenades (x,y) del centroide del cluster $i=1...3$.

La funció _cluster_data_ que us facilitem a continuació s'encarrega d'assignar les observacions de la matriu de dades X al clúster més proper d'una solució $S$. Específicament, la funció pren una solució com a entrada i retorna les magnituds cluster_centers, all_clusters_dists, cluster_indices, clusters, clusters_sum_dist. La funció fa servir les variables globals num_clusters ($k=3$), feature_vector_length ($d=2$) i X (matriu de dades a les que volem aplicar l'agorisme, en el nostre cas de dimensions $N \times d = 1500 \times 2$).

**NOTA IMPORTANT**: Tots els codis però hauran de ser escrits de forma que es puguin executar per un problema amb un nombre arbitrari de dimensions $d$ i de clústers $k$.


```python
def euclidean_distance(X, Y):
    return numpy.sqrt(numpy.sum(numpy.power(X - Y, 2), axis=1))

def cluster_data(solution):
    global num_clusters, feature_vector_length, X
    cluster_centers = []
    all_clusters_dists = []
    clusters = []
    clusters_sum_dist = []

    for clust_idx in range(num_clusters):
        cluster_centers.append(solution[feature_vector_length*clust_idx:feature_vector_length*(clust_idx+1)])
        cluster_center_dists = euclidean_distance(X, cluster_centers[clust_idx])
        all_clusters_dists.append(numpy.array(cluster_center_dists))

    cluster_centers = numpy.array(cluster_centers)
    all_clusters_dists = numpy.array(all_clusters_dists)

    cluster_indices = numpy.argmin(all_clusters_dists, axis=0)
    for clust_idx in range(num_clusters):
        clusters.append(numpy.where(cluster_indices == clust_idx)[0])
        if len(clusters[clust_idx]) == 0:
            clusters_sum_dist.append(0)
        else:
            clusters_sum_dist.append(numpy.sum(all_clusters_dists[clust_idx, clusters[clust_idx]]))

    clusters_sum_dist = numpy.array(clusters_sum_dist)

    return cluster_centers, all_clusters_dists, cluster_indices, clusters, clusters_sum_dist
```



a) [1 punt] Expliqueu breument a què corresponen cadascuna de les magnituds que retorna la funció _cluster_data_

- cluster_centers: És una llista de dos dimensions per emmagatzemar la posició dels centroides de la solució aplicada. Té la mida "num_clusters x feature_vector_length" (en l'exemple que estem fent servir, serà de mida 3x2).

- all_clusters_dists: És una llista de dos dimensions que serveix per emmagatzemar la distància dels centroides a cadascun dels punts de mostra de la matriu de dades. Té la mida "num_clusters x N" (en l'exemple que estem fent servir, serà de mida 3x1500).

- cluster_indices: Llista que emmagatzema l'índex del centroide més proper a cada mostra de de la matriu de dades. Per tant, té mida N (en el nostre cas 1500).

- clusters: Llista que agrupa els clusters (és a dir de cada centroide, les mostres que queden més properes i que formaran el cluster). En el nostre cas tindrem un llistat de 3xmostres de cada cluster.

- clusters_sum_dist: Llista que emmagatzema la suma de les distàncies dels centroides als seus clusters. Serà de mida "num_clusters" (en el nostre cas, 3).

b) [1 punt] Per formular el problema d'agrupament com un problema d'optimització el primer que cal entendre és que una solució òptima presentarà
un valor mínim de la distància euclídea entre cadascuna de les observacions $\vec{x}_j, j = 1,..N$ i el centroide del clúster al que han estat assignades $D$. Creeu una funció de fitness definida com l'invers de la magnitud $D$ i que serà la que haureu de maximitzar fent servir algorismes genètics.

```python
def f_fitness(solution):

    # Totes les variables innecessàries les marquem amb _ per optimitzar
    _, _, _, _, clusters_sum_dist = cluster_data(solution)

    # Afegim una delta baixa pels casos que la suma pugui ser 0
    # Al realitzar la divisió, obtenim un resultat mínim en el valor de la funció fitness
    fitness = 1.0 / (numpy.sum(clusters_sum_dist) + 0.000001)

    # Retornem una tupla
    return fitness,
```

c) [1 punt] Feu servir les llibreries DEAP per inicialitzar un algorisme genètic que maximitzi la funció de fitness definida a l'apartat anterior i en el que els individus són solucions $D$ del problema d'agrupament. Els detalls d'aquest algorisme són éls següents:

- Feu servir una inicialització aleatòria dels individus amb un valor entre -10 i +10.
- Apliqueu una estratègia de mutació gaussiana amb mitja $\mu = 0$ i desviació típica $\sigma = 5$ amb una probabilitat de mutació de 0.2.
- El mètode de selecció serà el millor entre 3 escollits aleatòriament
- El mètode d'encreuament en un punt

Consulteu la documentació de les llibreries DEAP per concretar els detalls de la implementació https://deap.readthedocs.io/en/master/

Per instal·lar i carregar les llibreries DEAP a l'entorn google colab executeu el codi següent

```python
!pip install deap
from deap import creator, base, tools, algorithms
```

```python
import random

# Definim que volem assolir l'objectiu de maximitzar la funció de fitness
creator.create ('FitnessMax', base.Fitness, weights = ( 1.0, ) )
creator.create ('Individual', list , fitness = creator.FitnessMax )

toolbox = base.Toolbox()
# Definim els individus i la població:
# Definim una inicialització aleatòria d'entre -10 i +10
toolbox.register ('attrFloat', lambda:random.random()*20-10)
# Cada individual representarà un centroide amb mida: 3 clusters x 2 atributs = 6 en el cas actual
toolbox.register ('individual', tools.initRepeat , creator.Individual , toolbox.attrFloat , n=num_clusters*feature_vector_length)
# La població serà un llistat d'individuals:
toolbox.register ('population', tools.initRepeat, list , toolbox.individual)

# Avaluarem amb la funció fitness definida anteriorment
toolbox.register ('evaluate', f_fitness )
# Mètode d'encreuament en un punt:
toolbox.register ('mate', tools.cxOnePoint )
# Aplicarem una estratègia de mutació gaussiana amb mitja 0, desviació típica 5 i probabilitat de mutació de 0.2
toolbox.register('mutate', tools.mutGaussian , mu=0 , sigma =5, indpb = 0.2 )
# Escollirem els millor de 3:
toolbox.register('select', tools.selTournament , tournsize =3)
```

d) [1 punt] Executeu l'algorisme genètic definit a l'apartat anterior amb un total de 1000 generacions amb un tamany de població de 10 individus i representeu gràficament la evolució de la funció de fitness promit en cada generació.

```python

import matplotlib.pyplot as plt
import numpy

# Definim una població de mida 10
poblacio = toolbox.population (10)
# La variable NGEN defineix el total de les generacions que realitzarem:
NGEN = 1000

# Definim Statistics per a poder representar la evolució promig en cada generació
stats = tools.Statistics ( key=lambda ind : ind.fitness.values )
# Només registrem "avg" perquè és el que es demana però es podria registrar altres com std, min, max
stats.register("avg", numpy.mean)
logbook = tools.Logbook()

# A clusters_generacions
centers_generacions = []

for gen in range(NGEN):
  # Hem definit una probabilitat de creuament de cada progenitor de cxpb i probabilitat de mutació 0.2
  # Ho he definit d'aquesta manera perquè després d'algunes combinacions he vist que era un resultat que
  # en la gràfica no representava tants alts i baixos
  descendents = algorithms.varAnd ( poblacio , toolbox , cxpb =0.5 , mutpb=0.1)
  # Calculem l'adaptació de cada individual al problema i associem l'adaptació  amb els individuals:
  adaptacions = toolbox.map( toolbox.evaluate , descendents )
  for fit , ind in zip ( adaptacions , descendents ) :
      ind.fitness.values = fit
  # Substiuim la població anterior amb la nova
  poblacio = toolbox.select (descendents , k=len(poblacio))

# Emmagatzemem els valors per després generar la gràfica:
  registre = stats.compile (poblacio)
  logbook.record (gen=gen, **registre)

  # Cada 100 generacions, guardem la informació dels centroides per representar a l'exercici 3C
  if (NGEN%100==0):
    centers_generacions.append(tools.selBest(poblacio, k=1)[0])

# Un cop ha fet les 1000 generacions, guardem els centroides resultants pel exercici 3A:
centers_result = tools.selBest(poblacio, k=1)[0]

# Recollim la informació de avg i de les generacions:
fitnessAvg = logbook.select ('avg')
generation = logbook.select ('gen')

# Dimensionem la gràfica
plt.figure(figsize=(15,10))

# Fiquem el text als exisos x i y
plt.xlabel ( "Generació" )
plt.ylabel ( "Adaptació")

# Dibuixem la evolució promig de cada generació
line = plt.plot (generation , fitnessAvg , "r-" , label = " Adaptació mitjana")

# Dibuixem la llegenda:
labs = [ l.get_label ( ) for l in line]
plt.legend (line, labs, loc = "center right" )

# Mostrem la gràfica
plt.show ( )
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib import animation

# Dimensionem la gràfica
fig = plt.figure(figsize=(15,10))
# Definim límits dels eixos x i y
ax = plt.axes(xlim=(-4,14),ylim=(-4,14))
# Indiquem el títol de la gràfica:
txt_title = ax.set_title('Evolució KMeans per iteració')

# Preparem la scatterplot amb la matriu de dades X
scatter1 =ax.scatter(X[:, 0], X[:, 1], s=40, cmap = 'viridis', edgecolors='black', label='Clústers')
# Preparem la scatterplot buit per dibuixar a posteriori els centroides corresponents
scatter2 =ax.scatter([], [], s=100, c="r", marker='X', label='Centroides')

# Dibuixem la llegenda
plt.legend(['Clústers','Centroides'], loc="upper left", prop={'size': 15});

# Primera iteració per conéixer el nombre necessari d'iteracions necessàries per convergir
iter_centers = 'random'
# Definim random_state=2 per a obtenir uns centroides inicials bastant random i així necessitar més iteracions
# num_clusters hem definit al exercici 1 que és 3 (k=3)
n_iteracions = KMeans(n_clusters=num_clusters, init=iter_centers, random_state=2)
# Apliquem KMeans definit a la matriu de dades X
n_iteracions.fit(X)
# Obtenim el nombre d'iteracions realitzades per convergir a max_iterations
max_iterations = n_iteracions.n_iter_

# Estructures per emmagatzemar els clústers i els labels de les diferents iteracions:
clusters_iteracions=[]
labels_iteracions=[]

# Itero fins al màxim de iteracions definides
for i in range (1, max_iterations+1):
  # Definim KMeans que volem executar indicant:
  # n_clusters: nombre de clusters que ve definit com a variable num_clusters en exercicis anteriors
  # init: L'array de centroides obtinguts en la iteració anterior per informar-la com a inici de la iteració actual (el primer cop serà 'random')
  # max_iter: Farem una única iteració (tenint en compte que partim de la situació de la iteració anterior)
  # random_state=2: D'aquesta manera generem centroides inicials més random per poder dibuixar millor la evolució
  kmeans = KMeans(n_clusters=num_clusters, init='random', n_init=1, max_iter=i, random_state=2)
  # Aplico l'algorisme KMeans definit a la matriu de dades generades a l'exercici 1, X
  kmeans.fit(X)
  # Emmagatzemo els centroides generats a iter_centers i els labels a partir de la predicció a iter_labels
  iter_centers = kmeans.cluster_centers_
  iter_labels = kmeans.predict(X)
  # Es guarden tant els centroides com els labels de cada iteració en les estructures creades anteriorment
  clusters_iteracions.append(iter_centers)
  labels_iteracions.append(iter_labels)

# Funció per anar actualitzant la gràfica.
# El paràmetre 'n' correspon al número de la iteració
# El paràmetre 'iter_centers' correspon al llistat de centroides emmagatzemats en cada iteració
# El paràmetre 'iter_labels' correspon al llistat de labels emmagatzemats en cada iteració
def animate(n, iter_centers, iter_labels):

    # Actualitzem a cada iteració les dades de la classificació de cada cluster (iter_labels) que li correspon a la iteració 'n'
    #scatter1.set_cmap('viridis')
    scatter1.set_array(iter_labels[n])
    # Actualitzem a cada iteració les dades de la classificació de cada centroide (iter_centers) que li correspon a la iteració 'n'
    scatter2.set_offsets(iter_centers[n])

    # Inform the position of cluster 1
    #annotation1.set_position((-iter_centers[n][0,0],-iter_centers[n][0,1]))
    #annotation1.xy = (-iter_centers[n] [0,0],-iter_centers[n][0,1])

    # Indiquem títol a la gràfica per indicar quina iteració es mostra
    txt_title.set_text('Iteració = {0:4d}'.format(n))
    # Indiquem les posicions de cada centroide
    centroide1 = 'Centroide1: [' + str((round(iter_centers[n][0, 0]), 2)) + ", " + str((round(iter_centers[n][0, 1]), 2)) + ']'
    plt.text(-1, 6.5, centroide1, fontsize=15, color='black')
    centroide2 = 'Centroide2: [' + str((round(iter_centers[n][1, 0]), 2)) + ", " + str((round(iter_centers[n][1, 1]), 2)) + ']'
    plt.text(-1, 7.5, centroide2, fontsize=15, color='black')
    centroide3 = 'Centroide3: [' + str(round((iter_centers[n][2, 0]), 2)) + ", " + str((round(iter_centers[n][2, 1]), 2)) + ']'
    plt.text(-1, 8.5, centroide3, fontsize=15, color='black')

    return (scatter1,scatter2)

# Fem servir FuncAnimation per veure l'evolució dels centroides
# Paràmetre "frames": li passem un integer, és equivalent a passar: range(max_iterations). Permet simular el bucle per dibuixar la gràfica.
# Paràmetre "interval": definim amb 3000 milisegons per poder veure la evolució iteració a iteració
# Paràmetre "fargs": Seran els arguments que li passarem a la funció animate() per dibuixar l'evolució a partir de les iteracions.
anim = animation.FuncAnimation(fig, animate, frames=max_iterations, interval=3000, fargs=[clusters_iteracions, labels_iteracions])

#Generem un vídeo per reproduir les gràfiques
from IPython.display import HTML
HTML(anim.to_html5_video())
```
