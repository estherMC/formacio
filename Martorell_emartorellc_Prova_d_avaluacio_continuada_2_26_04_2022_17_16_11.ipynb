```python
# 1.f: afegiu i executeu el vostre codi aquí

import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA

# Apliquem PCA (considerant tots els atributs) a les dades processades (escalades)
pca = PCA().fit(data_processed)

# Calculem variància acumulada
cumsum_list = [np.concatenate([[0],pca.explained_variance_ratio_.cumsum()])]
df_cumsum = pd.DataFrame(np.array(cumsum_list).T)
df_cumsum.columns = ['PCA']

# Dibuixem gràfica
fig, ax = plt.subplots(1,1,figsize=(10, 6), sharex=True)
df_cumsum.plot(ax=ax, color=sns.color_palette('Set1',5)[3:5],
               kind='line', marker='o')
ax.axhline(y = 0.90, color='k', linestyle='--', label = '90% Variança')
ax.grid()
plt.title('Variança acumulada explicada en funció del número de components',
          fontsize='x-large')
ax.set_ylabel('Variança')
ax.set_xlabel('Components PCA')
ax.set_xlim([0,len(cumsum_list[0])-0.5])
ax.set_ylim([0,1.05])
ax.legend(loc='lower right')

# Mostrem gràfica
plt.show()

print('Els components necessaris per representar el 90% de la variança són: ' +
      str(sum(df_cumsum['PCA']<0.90)))
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine

raw_data = load_wine()

df_features = pd.DataFrame(raw_data['data'],
                           columns=raw_data['feature_names'])

df_labels = pd.DataFrame({'label': raw_data['target']})

df_labels[df_labels==2] = 0
```

**2.a) (1 PUNT)** Entreneu un arbre de decisió de profunditat 4, implementat en _scikit-learn_, fent servir el 70% de les dades disponibles (conjunt d’entrenament) i calculeu el temps d’entrenament

https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

```python
# 2.a: afegiu i executeu el vostre codi aquí
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import timeit

# Dividir el conjunt de dades fent servir el 70%
X_train, X_test, y_train, y_test = train_test_split(df_features, df_labels, train_size=0.7, random_state=0)

# Creem un arbre de decisió de profunditat 4
dt_classifier = DecisionTreeClassifier(max_depth=4)

# Entrenem i calculem el temps d'entrenament
starttime = timeit.default_timer()
tree_fit = dt_classifier.fit(X_train, np.ravel(y_train))
print("El temps d'entrenament de DecisionTreeClassifier amb profunditat 4 és:", timeit.default_timer() - starttime)
```

**2.c) (1 PUNT)** Per el conjunt de test resultant (30% de les dades), representeu la matriu de confusió. Proporcioneu una definició amb les vostres paraules (no us limiteu a la fòrmula) de les mètriques precisió, recall i F1-score, i obtingueu els seus valors en el conjunt de test per totes les classes.

```python
# 2.c: afegiu i executeu el vostre codi aquí
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import confusion_matrix

# S'obtenen les prediccions a partir de les dades test
predicted_labels = dt_classifier.predict(X_test)

# Dibuixem la matriu de confusió
sns.heatmap(confusion_matrix(y_true=y_test, y_pred=predicted_labels), annot = True, fmt=".0f", annot_kws={"size":18})

# Indiquem els noms dels eixos i el títol de la matriu
plt.xlabel('Prediccions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Matriu Confusió', fontsize=18)
# Mostrem la matriu
plt.show()

# Es calculen els valors de precisió, recall i F1 indicant els valors de y_test reals
# i comparant-los amb els valors que s'han predit anteriorment (predicted_labels)
precision = metrics.precision_score(y_test, predicted_labels)
recall = metrics.recall_score(y_test, predicted_labels)
f1 = metrics.f1_score(y_test, predicted_labels)

print("Valor de precisió és: {:.2f}".format (precision))
print("Valor de recall: {:.2f}".format (recall))
print("Valor de F1: {:.2f}".format (f1))
```

La matriu de confusió permet conèixer i avaluar totes les combinacions de quins valors s'han mesurat d'una determinada forma i com són en la realitat. Els tipus que s'obtenen s'anomenen: "True Negative" (TN), "True Positive" (TP), "False Positive" (FP) i "False Negative" (FN). Nota: Feu servir _sklearn.model_selection.StratifiedKFold_ en el paràmetre _cv_ de _cross_validate_ per crear els subconjunts (_folds_, _splits_).

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedKFold.html

```python
# 2.d: afegiu i executeu el vostre codi aquí
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold

# Entrenem un classificador AdaBoost amb arbre de profunditat 2
adaboost_clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth=2))

# Creem llistat de mètriques que volem obtenir
metric_names = ['precision', 'recall', 'f1']
# Creem 5 subconjunts
skf = StratifiedKFold(n_splits=5)

scores = []

# Realitzem la validació creuada a tot el conjunt de dades
scores = cross_validate(adaboost_clf, df_features, np.ravel(df_labels), cv=skf, scoring=metric_names)

# Imprimim les mètriques obtingudes
print ("Mean:               - Precision: {:.2f} - Recall: {:.2f} - F1: {:.2f}".
  format(scores['test_' + metric_names[0]].mean(), scores['test_' + metric_names[1]].mean(), scores['test_' + metric_names[2]].mean()) )
print ("Standard deviation: - Precision: {:.2f} - Recall: {:.2f} - F1: {:.2f}".
  format(scores['test_' + metric_names[0]].std(), scores['test_' + metric_names[1]].std(), scores['test_' + metric_names[2]].std()))
```

Observem com a partir dels valors de la mitjana obtinguts, tant la precisió com la mesura Recall tenen valors molt elevats (pròxims al 1). A conseqüència d'aquestes dues mesures, el valor de F1 té un valor molt elevat també. Aquest valor de F1 ens determina que amb el classificador AdaBoost amb un arbre de profunditat 2 obtenim molt bons resultats.

La desviació estàndard també va lligada a uns bons resultats ja que al disposar d'una desviació estàndard tant baixa, significa que la major part de les dades de mostra tendeixen a estar agrupats aprop de la seva mitjana.
